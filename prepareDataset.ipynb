```python
# Setting other paths for the dataset 
typeA = "typeA/TypeA.jpg" 
typeB = "typeB/TypeB.jpg" 
typeC = "typeC/TypeC.jpg" 
typeD = "typeD/TypeD.jpg" 
typeE = "typeE/TypeE.jpg" 
typeF = "typeF/TypeF.jpg" 
```

```python
# Creating a function for creating multiple images 
def create_multiple_images(image_path, numbers, typeTag): 
    # Loading the image into memory, and resize the image by 
    # 98x98 
    image = cv2.imread(image_path)
    image = cv2.resize(image, (98, 98))
    
    # Convert the image into a gray scale image 
    image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY) 
    
    # For ranges of values from zero to 100, apply a threshold value with 
    # each respective count and apply it into the loaded image 
    for count in range(0, numbers): 
        # Get the threshold image 
        ret, thresh_image = cv2.threshold(image, count, 255, cv2.THRESH_BINARY)
        
        # Save the image 
        file_name = f"type{typeTag}{count}.jpg"
        cv2.imwrite(file_name, thresh_image) 
        
```

```python
# Creating images for TypeA 
# Creating 250 sample images 
numbers = 250
create_multiple_images(typeA, numbers, "A") 
```
